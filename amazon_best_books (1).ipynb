```python
names, authors, ratings, reviews, prices =[],[],[],[],[]
for pg in range(1,3):
    url=f"https://www.amazon.in/gp/bestsellers/books/ref=zg_bs_pg_{pg}?ie=UTF8&pg={pg}"
    page=requests.get(url)
    soup=BeautifulSoup(page.content, 'lxml')
    books=soup.find_all(class_="zg-grid-general-faceout")
    for p in books:
        det=[t.get_text() for t in p.find_all(class_="_cDEzb_p13n-sc-css-line-clamp-1_1Fn1y")]
        if len(det)==2:
            names.append(det[0])
            authors.append(det[1])
        else:
            det=[t.get_text() for t in p.find_all(class_="_cDEzb_p13n-sc-css-line-clamp-2_EWgCb")]
            names.append(det[0])
            authors.append(None)
        if p.find(class_="a-icon-row"):
            ratings.append(p.find(class_="a-icon-row").a['title'])
            reviews.append(p.find(class_="a-icon-row").find_all('span')[-1].get_text())
        else:
            ratings.append(None)
            reviews.append(None)
        
        prices.append(p.find(class_="p13n-sc-price").get_text())

len(names), len(authors), len(ratings), len(reviews), len(prices)        
```

```python
d={"Book Name":names,'Authors':authors,'Ratings':ratings, "Reviews":reviews,'Prices':prices}
df=pd.DataFrame(d)
df
```
